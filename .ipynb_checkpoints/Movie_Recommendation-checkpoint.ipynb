# Creating Movie Recommendation

```python
df.head()
```

For every user, we will mark his/her rating in the movie column, if not rated then NaN

```python
movie_matrix = df.pivot_table(index = "user_id", columns = "title", values = "rating")
```

```python
movie_matrix
```

```python
ratings.sort_values('Number of Ratings', ascending = False).head()
```

```python
starwars_user_rating = movie_matrix['Star Wars (1977)']
starwars_user_rating.head()
```

```python
# How much correlated starwars is with other movies.. 
similar_to_starwars = movie_matrix.corrwith(starwars_user_rating)
```

```python
# if NaN in movie X, then that person has not rated starwars and X
```

```python
corr_starwars = pd.DataFrame(similar_to_starwars, columns=["Correlation"])
```

```python
corr_starwars.dropna(inplace=True)
```

```python
corr_starwars.head()
```

```python
corr_starwars.sort_values('Correlation', ascending = False).head(10)
```

```python
ratings
```

```python
corr_starwars = corr_starwars.join(ratings['Number of Ratings'])
```

```python
corr_starwars.head()
```

```python
corr_starwars[corr_starwars['Number of Ratings']>100].sort_values('Correlation', ascending = False)
```

## PREDICT FUNCTION

```python
def predict_movies(movie_name):
    # Make the Movie Matrix
    movie_user_rating = movie_matrix[movie_name]
    
    # Find Correlation of that movie with others
    similar_to_movie = movie_matrix.corrwith(movie_user_rating)
    
    # Make Data Frame
    corr_movie = pd.DataFrame(similar_to_movie, columns=["Correlation"])
    corr_movie.dropna(inplace=True)
    
    # join
    corr_movie = corr_movie.join(ratings['Number of Ratings'])
    
    #sort with filtering
    predictions = corr_movie[corr_movie['Number of Ratings']>100].sort_values('Correlation', ascending = False)
    
    return predictions
```

```python
predictions = predict_movies('Titanic (1997)')
```

```python
predictions.head(10)
```

```python
predictions = predict_movies('Men in Black (1997)')
```

```python
predictions.head(10)
```

```python
predictions = predict_movies('Tomorrow Never Dies (1997)')
```
